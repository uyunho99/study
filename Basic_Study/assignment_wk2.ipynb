### Problem 1

어떤 벡터가 주어졌을 때 L2 norm을 구하는 함수 <code>get_L2_norm()</code>을 작성하세요.

- 매개변수 : 1차원 벡터 (np.array)
- 반환값 : 인자로 주어진 벡터의 L2 Norm값 (number)

```python
import numpy as np

mat = np.array([1, 2, 3, 4])

def get_L2_norm(mat):
    return np.linalg.norm(mat, 2)

get_L2_norm(mat)
```

### Problem 2

어떤 행렬이 singular matrix인지 확인하는 함수 <code>is_singular()</code>를 작성하세요.

- 매개변수 : 2차원 벡터(np.array)
- 반환값 : 인자로 주어진 벡터가 singular하면 True, non-singular하면 False를 반환

```python
import numpy as np

mat_1 = np.array([[1,2],[3,4]])
mat_2 = np.array([[1,2],[2,4]])

def is_singular(mat):
    return True if np.linalg.det(mat) == 0 else False

print(is_singular(mat_1))
print(is_singular(mat_2))
```
