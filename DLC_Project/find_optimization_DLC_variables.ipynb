```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
import shutil
import os
```

```python
def load_DLC_project(project_name, iter_num):

    # Get project directory name
    dir_name = "/mnt/sdb2/DeepDraw/DLC_Project/{}".format(project_name)
    if not os.path.isdir(dir_name):
        print("The project does not exist, please create the project")
    
    info_path = os.path.join(dir_name, "evaluation-results/iteration-{}/CombinedEvaluation-results.csv".format(iter_num))
    info = pd.read_csv(info_path)
    
    return dir_name, info_path, info
```
